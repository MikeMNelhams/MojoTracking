```python
import matplotlib.pyplot as plt
from skimage.feature import hog
from skimage import data, color, exposure
import numpy as np
import cv2
```

```python
## First frame of video cover0_0
image = cv2.imread(r"C:\Users\Shane\Desktop\Year 3\Mathematical and Data Modelling\Phase B\MDM3_B\OrientationJ_Data\1st source\Video 0\Frames\frame0.jpg")
```

```python
#First bounding box for that frame -- if run close picture image to continue
crop = image[410:410+64, 642:642 + 61]
cv2.imshow("cropped", crop)
cv2.waitKey(0)
```

```python
## Converting to histogram of oriented gradients image

fd, hog_image = hog(crop, orientations=8, pixels_per_cell=(2,2),
                    cells_per_block=(1,1), visualize=True, multichannel=True)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4), sharex=True, sharey=True)

ax1.axis('off')
ax1.imshow(crop, cmap=plt.cm.gray)
ax1.set_title('Input image')

# Rescale histogram for better display
hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 40))

ax2.axis('off')
ax2.imshow(hog_image_rescaled, cmap=plt.cm.gray)
ax2.set_title('Histogram of Oriented Gradients')
plt.show()
```
